```python
#data cleaning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 2. Create a Dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Eve', None],
    'Age': [20, 21, -5, 21, 200, 22],
    'Marks': [85, np.nan, 78, 78, 95, 300],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', '']
}
df = pd.DataFrame(data)
df

# •3. Handling Missing Values
df.isnull().sum()
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df['Name'] = df['Name'].fillna('Unknown') 
# •4. Removing Duplicate Records
df = df.drop_duplicates(subset=['Name', 'Age', 'City'])
#5. Standardize text
df['Name'] = df['Name'].str.strip().str.title()
df['City'] = df['City'].str.strip().str.title()
#.str - Removes leading and trailing whitespace from each string.
#.title() - Converts the first letter of each word to uppercase, the rest to lowercase.
#6. Handle invalid
df.loc[(df['Age'] < 0) | (df['Age'] > 120), 'Age'] = np.nan
df['Age'] = df['Age'].fillna(df['Age'].median())
#7. Handle missing marks and Outliers 
df.loc[(df['Marks'] < 0) | (df['Marks'] > 100), 'Marks'] = np.nan
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Sample data
df = pd.DataFrame({'Age': [15, 22, 25, 30, 35, 40, 50, 60, 70, 80]})
# Max-min/number of bins 
# 80-15/3=65/3=21.67
# 15+21.67= 15-36.67
# Equal-Width Binning
df['Equal-Width Binning'] = pd.cut(df['Age'], bins=3, labels=['Young', 'Middle-Aged', 'Old'])
print(df)
# Equal-Frequency Binning
df['Equal-Frequency Binning'] = pd.qcut(df['Age'], q=3, labels=['Low', 'Medium', 'High'])
print(df)
# Custom Binning
bins = [0, 18, 35, 50, 100]
labels = ["Teen", "Young Adult", "Adult", "Senior"]
df['Custom Binning'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df)
```

```python
# Plotting
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
# Equal-Width Binning Bar Chart
sns.countplot(x=df["Equal-Width Binning"], 
              palette="Blues", ax=axes[0])
axes[0].set_title("Equal-Width Binning")
# Equal-Frequency Binning Bar Chart
sns.countplot(x=df["Equal-Frequency Binning"], palette="Greens", ax=axes[1])
axes[1].set_title("Equal-Frequency Binning")
# Custom Binning Bar Chart
sns.countplot(x=df["Custom Binning"], palette="Oranges", ax=axes[2])
axes[2].set_title("Custom Binning")
# Show plot
plt.tight_layout() #Matplotlib that automatically adjusts the spacing between subplots to prevent overlapping.
plt.show()
```

```python
# Market Basket Analysis – Full Merged Runnable Program

# Install library (run in terminal if required)
# pip install mlxtend

import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# -------------------------------
# Sample Market Basket Data (First Example)
# -------------------------------
transactions = [
    ['Milk', 'Bread'],
    ['Milk', 'Diapers', 'Beer'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Butter'],
    ['Milk', 'Juice']
]

# Convert transactions to One-Hot Encoding format
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

print("Dataset (One Hot Encoded):")
print(df)

# -------------------------------
# Frequent Itemset Mining (Apriori)
# -------------------------------
frequent_itemsets = apriori(
    df,
    min_support=0.4,
    use_colnames=True
)

print("\nFrequent Itemsets:")
print(frequent_itemsets)

# -------------------------------
# Association Rules Generation
# -------------------------------
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# -------------------------------
# Strong Rule Filtering
# -------------------------------
strong_rules = rules[
    (rules['confidence'] >= 0.7) &
    (rules['lift'] > 1)
]

print("\nStrong Rules:")
print(strong_rules[['antecedents', 'consequents', 'confidence', 'lift']])

# -------------------------------
# Second Dataset (Assignment Dataset)
# -------------------------------
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Beer', 'Bread'],
    ['Milk', 'Bread', 'Beer', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Butter'],
    ['Beer', 'Butter'],
    ['Milk', 'Bread', 'Beer'],
    ['Milk', 'Butter']
]

# Prepare dataset
df = pd.DataFrame(transactions)

# One-hot encoding using pandas method
one_hot = pd.get_dummies(df.stack()).groupby(level=0).sum()

print("\nPrepared Dataset:")
print(one_hot)

# -------------------------------
# Apriori Algorithm (Assignment Requirement)
# -------------------------------
frequent_itemsets = apriori(
    one_hot,
    min_support=0.3,
    use_colnames=True
)

print("\nFrequent Itemsets (min_support=0.3):")
print(frequent_itemsets)

# -------------------------------
# Association Rules Generation
# -------------------------------
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.7
)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# -------------------------------
# Analysis Answers
# -------------------------------

# Most frequently bought item
item_frequency = one_hot.sum().sort_values(ascending=False)
print("\nMost Frequently Bought Item:")
print(item_frequency.head(1))

# Strong rule recommendation
if not rules.empty:
    best_rule = rules.sort_values(
        ['lift', 'confidence'],
        ascending=False
    ).iloc[0]

    print("\nRecommended Rule:")
    print(best_rule[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# Filter rules with lift > 1.2
strong_rules = rules[rules['lift'] > 1.2]

print("\nRules with Lift > 1.2:")
print(strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# -------------------------------
# Visualization
# -------------------------------
plt.figure(figsize=(6,4))
plt.scatter(rules['support'], rules['confidence'])
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Association Rules Scatter Plot")
plt.grid()
plt.show()
```

```python
# =========================================
# DMDA - Unit 3
# Classification and Regression
# =========================================

# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Independent variable (X)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)

# Dependent variable (Y)
Y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, Y)

print("Slope (m):", model.coef_)      # slope (m)

print("Intercept (c):", model.intercept_)  # intercept (c)

Y_pred = model.predict(X)

print("Predicted values:", Y_pred)

plt.scatter(X, Y, color='blue')   # Actual points
plt.plot(X, Y_pred, color='red')  # Regression line
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression Example")
plt.show()
```
